```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import re

data = pd.read_csv('data\\training_data_202601.csv')
feature_names = np.array(data.columns).tolist()
target_name = feature_names.pop(1)
print('Feature Names:', feature_names)
print('Target Name:', target_name)

data['This art piece makes me feel calm.'] = data['This art piece makes me feel calm.'].str.extract(r'(\d+)').astype(float)
data['This art piece makes me feel content.'] = data['This art piece makes me feel content.'].str.extract(r'(\d+)').astype(float)
data['This art piece makes me feel sombre.'] = data['This art piece makes me feel sombre.'].str.extract(r'(\d+)').astype(float)
data['This art piece makes me feel uneasy.'] = data['This art piece makes me feel uneasy.'].str.extract(r'(\d+)').astype(float)
numeric_features = data[feature_names].select_dtypes(include=[np.number]).columns.tolist()
print('Numeric Feature Names:', numeric_features)
```

```python
for feature in numeric_features[1:]:
    
    data[feature].hist(bins=30)
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.vlines(data[feature].mean(), ymin=0, ymax=data[feature].value_counts().max(), color='red', linestyle='dashed', linewidth=1, label='Mean value: {:.2f}'.format(data[feature].mean()))
    plt.legend()
    plt.show()
```

```python
feature_data = data[feature_names].to_numpy()
target_data = data[target_name].to_numpy()

print('Feature Data shape:\n', feature_data.shape)
print('Target Data shape:\n', target_data.shape)
```
